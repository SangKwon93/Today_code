```python
import pandas as pd
```

```python
file='C:/Users/psk/3rd_backup/Kream_pro.csv'
file1='C:/Users/psk/3rd_backup/trade_history.csv'
file2='C:/Users/psk/3rd_backup/kream_detail.csv'
```

```python
df = pd.read_csv(file)
df
```

```python
df3 = pd.read_csv(file2)
df3 = df3.loc[:,['_id','model_no']]
df3
```

```python
tm1 = pd.merge(df,df3, on='_id', how='left')
tm1
```

```python
df1 = pd.read_csv(file1,index_col=0)
df1
```

```python
# 제품별 거래횟수(235,270)
df2 = df1.groupby('model_no')['trade_date'].count().to_frame().reset_index()
df2 = df2.rename(columns={"trade_date":"count"})
df2[df2['model_no']=='555088-105']
```

```python
# 데이터 빈공간 삭제 함수
def strip_pro(data):
    data= data.strip()
    return data
```

```python
tm1['model_no']= tm1['model_no'].apply(strip_pro)
tm1
```

```python
k1 =pd.merge(tm1,df2,on='model_no',how="outer")
k1
```

```python
k1 = k1.fillna(0)
k1 = k1.astype({'count':'int'})
k1
```

```python
product_table = k1.loc[:,['model_no','_id','all_trade']]
product_table
```

```python
product_table.to_csv('C:/Users/psk/3rd_backup/ERD_table/trade_table.csv', index=False, encoding='cp949')
```
