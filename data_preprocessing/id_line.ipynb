```python
import pymongo
import pandas as pd
import re
```

```python
file='C:/Users/psk/3rd_backup/Kream_detail.csv'
```

```python
df = pd.read_csv(file,index_col=0)
df.head()
```

```python
df = df.loc[:,['_id','model_no']]
df.head()
```

```python
data1= '(GS) Jordan 1 Retro High Melo'
```

```python
# 정규식을 활용
p = re.compile('Jordan \d+', re.M)
result = p.findall(data1)
# result = [i[2:] for i in result]
result
```

```python
def extract_line(data):
    p = re.compile('Jordan \d+', re.M)
    result = p.findall(data)
    # return result[0]
    result = [i[:] for i in result]
    return ','.join(result)
    
```

```python
extract_line('Jordan 1 Retro High OG Bleached Coral	')
```

```python
df['line'] = df['_id'].apply(extract_line)
df
```

```python
df = df.loc[:,["model_no",'line']]
df.head()
```

```python
def strip_pro(data):
    data = data.strip()
    return data
```

```python
# 컬럼 model_no에 공백이 있어 join이 불가하여 strip_pro 함수 적용
df['model_no']= df['model_no'].apply(strip_pro)
df
```

```python
df.to_csv('kream_detail_line.csv', index=False, encoding='cp949')
```
