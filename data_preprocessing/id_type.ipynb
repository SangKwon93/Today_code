```python
import pymongo
import pandas as pd
import re
```

```python
file='C:/Users/psk/3rd_backup/Kream_detail.csv'
```

```python
df = pd.read_csv(file,index_col=0)
df.head()
```

```python
df = df.loc[:,['_id','model_no']]
df.head()
```

```python
list(df['_id'])
```

```python
# Type에 대한 정보 컬럼 추가
def extract_type(data):
    if 'High' in data:
        return 'High'
    elif 'Mid' in data:
        return 'Mid'
    elif 'Low' in data:
        return 'Low'
    else:
        return 'Others'
```

```python
extract_type('Jordan 1 x A Ma Maniere Retro High OG SP Sail and Burgundy')
```

```python
df['type'] = df['_id'].apply(extract_type)
df = df.loc[:,["model_no",'type']]
df.head()
```

```python
def strip_pro(data):
    data = data.strip()
    return data
```

```python
# 컬럼 model_no에 공백이 있어 join이 불가하여 strip_pro 함수 적용
df['model_no']= df['model_no'].apply(strip_pro)
df
```

```python
df.to_csv('kream_detail_type.csv', index=False, encoding='cp949')
```
