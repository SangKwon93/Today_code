```python
# all trade 전처리
# 거래 8.8만 -> 88000
def num_change(data):
    data = data[3:]
    if data == '':
        data = '0'
    if '.' in data:
        data = data.replace('만', '000')
    else:
        data = data.replace('만', '0000')
    data = data.replace('.', '').replace(',', '')
    data = int(data)
    
    return data
```

```python
# price 전처리
# 600,000원 -> 600000
def change_price(data):
    data = data.replace('-', '0')
    data = int(data.replace('원', '').replace(',', ''))
    return data
```
